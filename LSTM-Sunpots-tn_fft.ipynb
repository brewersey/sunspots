```python
#creating X and y data
```

```python
y = sunspots_month['13-Month Smoothed Total Sunspot Number']
y.dropna(inplace = True)
y.head()
```

```python
X = [x for x in range(0, len(y))]
print(X)
```

```python
plt.plot(X, y)
plt.show()
```

```python
#FFT of sunspot data
```

```python
sp_fft = np.fft.fft(y)
```

```python
plt.plot(abs(sp_fft))
plt.xlim(0, 200)
plt.show()
```

```python
n = len(sp_fft)
power = abs(sp_fft[1:(n//2)])**2
nyquist = 1.0/2
freq = np.array(range(n//2))/(n//2.0)*nyquist
period = 1.0/freq[1:]/12
```

```python
plt.plot(period, power)
plt.xlim(0, 45)
plt.show()
```

```python
spot_cycle = period[np.argmax(power)]
spot_cycle
```
